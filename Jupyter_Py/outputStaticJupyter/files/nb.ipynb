```python
from ipyleaflet import *

# 1. Definisci i due layer (Sinistro e Destro)
left_layer = basemap_to_tiles(basemaps.OpenStreetMap.Mapnik)
right_layer = basemap_to_tiles( basemaps.Esri.WorldImagery ) #basemaps.Esri.WorldImagery)

# 2. Crea la mappa base
m = Map(center=(41.89, 12.49), zoom=12)

# 3. Crea e aggiungi il controllo SplitMap
split_control = SplitMapControl(left_layer=left_layer, right_layer=right_layer)
m.add_control(split_control)

m
```
